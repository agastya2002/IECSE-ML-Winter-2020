```python
import pandas as pd
import matplotlib.pyplot as plt

obj =  pd.read_csv("Sample.csv", names = ['x','y'])
```

```python
plt.figure(figsize = (14,8))
X = obj['x']
Y = obj['y']
plt.plot(X,Y,color = "red")
plt.grid()
plt.title("Epic Info")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
