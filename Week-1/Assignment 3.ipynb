```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sample.csv",header=None)
print(df)
```

```python
x = np.array(df.iloc[:,0])
y = np.array(df.iloc[:,1])

fig,ax=plt.subplots()
ax.set(xlabel='X Axis',ylabel='Y Axis',title='Epic Info')
ax.plot(x,y,'r')
ax.grid()

plt.rcParams.update({'font.size': 15})
plt.show()
```
